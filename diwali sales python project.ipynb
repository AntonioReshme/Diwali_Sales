```python
# import python libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# import csv file

df = pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
# Cleaning the data

df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#drop unrelated/blank columns

df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
#check for null values

df.isnull().sum()
```

```python
# drop null values

df.dropna(inplace=True)
```

```python
df.shape
```

```python
# change data type

df["Amount"] = df["Amount"].astype(int)
```

```python
df.dtypes
```

```python
df.columns
```

```python
#rename column

df.rename(columns={"Marital_Status":"Marriage"})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)

df.describe()
```

```python
# use describe() for specific columns

df[["Age","Orders","Amount"]].describe()
```

```python
# Exploratory Data Analysis
# plotting a bar chart for Gender and it's count

ax = sns.countplot(data=df,x="Gender",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(data=sales_gen,x="Gender",y="Amount",hue="Gender")
```

```python
# From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
# Age

ax = sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group

sales_age = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(["Amount"],ascending=False)
sns.barplot(data=sales_age,x="Age Group",y="Amount")
```

```python
# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
# total number of orders from top 10 states

order_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(["Orders"],ascending=False).head(10)
sns.barplot(data=order_state,x="State",y="Orders")
sns.set(rc={"figure.figsize":(20,5)})
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(["Amount"],ascending=False).head(10)
sns.barplot(data=sales_state,x="State",y="Amount")
sns.set(rc={"figure.figsize":(20,5)})
```

```python
# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
df.info()
```

```python
# Marital Status

ax = sns.countplot(data=df,x="Marital_Status",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
mari_sales = df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(["Amount"],ascending=False)
sns.barplot(data=mari_sales,x="Marital_Status",y="Amount",hue="Gender")
```

```python
# From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
# Occupation

ax = sns.countplot(data=df,x="Occupation")
sns.set(rc={"figure.figsize":(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occ_amount = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(["Amount"],ascending=False)
sns.barplot(data=occ_amount,x="Occupation",y="Amount")
```

```python
# From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
# Product Category

ax = sns.countplot(data=df,x="Product_Category")
sns.set(rc={"figure.figsize":(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_sales = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(["Amount"],ascending=False).head(10)
sns.barplot(data=product_sales,x="Product_Category",y="Amount")
sns.set(rc={"figure.figsize":(20,5)})
```

```python
# From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
# top 10 most sold products (same thing as above)
total_orders = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(["Orders"],ascending=False).head(10)
sns.barplot(data=total_orders,x="Product_ID",y="Orders")
```
